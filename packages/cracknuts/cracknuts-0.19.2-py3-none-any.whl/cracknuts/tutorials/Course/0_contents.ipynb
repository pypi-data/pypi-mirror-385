### 分析代码说明

本教程分析部分的代码fork自scarr库（https://github.com/decryptofy/scarr.git ），
地址为 https://github.com/cracknuts-team-admin/scarr，
我们会根据需要进行功能添加，并尽量贡献到scarr库中。

使用本教程之前，先从我们fork的地址安装scarr库： 
```
pip3 install "git+https://github.com/cracknuts-team-admin/scarr.git"
```
